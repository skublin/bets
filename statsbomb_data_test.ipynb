```python
df_selected_events = df_events[['timestamp', 'team', 'type', 'minute', 'location', 'pass_end_location', 'player']]
df_selected_events = df_selected_events.sort_values(['minute', 'timestamp'])
```

```python
# let's divide events on two teams
df_arg_events = df_selected_events[df_selected_events['team'] == 'Argentina']
df_fra_events = df_selected_events[df_selected_events['team'] == 'France']
```

```python
# now get pass events only
df_arg_pass = df_arg_events[df_arg_events['type'] == 'Pass']
df_fra_pass = df_fra_events[df_fra_events['type'] == 'Pass']
```

```python
# let's move this drawing to one function 
def show_passes(df_passes):
    # drawing pitch based on https://fcpython.com/visualisation/drawing-pass-map-python

    #Create figure
    fig=plt.figure()
    fig.set_size_inches(15, 10)
    ax=fig.add_subplot(1,1,1)

    #Pitch Outline & Centre Line
    plt.plot([0,0],[0,90], color="black")
    plt.plot([0,130],[90,90], color="black")
    plt.plot([130,130],[90,0], color="black")
    plt.plot([130,0],[0,0], color="black")
    plt.plot([65,65],[0,90], color="black")

    #Left Penalty Area
    plt.plot([16.5,16.5],[65,25],color="black")
    plt.plot([0,16.5],[65,65],color="black")
    plt.plot([16.5,0],[25,25],color="black")

    #Right Penalty Area
    plt.plot([130,113.5],[65,65],color="black")
    plt.plot([113.5,113.5],[65,25],color="black")
    plt.plot([113.5,130],[25,25],color="black")

    #Left 6-yard Box
    plt.plot([0,5.5],[54,54],color="black")
    plt.plot([5.5,5.5],[54,36],color="black")
    plt.plot([5.5,0.5],[36,36],color="black")

    #Right 6-yard Box
    plt.plot([130,124.5],[54,54],color="black")
    plt.plot([124.5,124.5],[54,36],color="black")
    plt.plot([124.5,130],[36,36],color="black")

    #Prepare Circles
    centreCircle = plt.Circle((65,45),9.15,color="black",fill=False)
    centreSpot = plt.Circle((65,45),0.8,color="black")
    leftPenSpot = plt.Circle((11,45),0.8,color="black")
    rightPenSpot = plt.Circle((119,45),0.8,color="black")

    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)

    #Prepare Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="black")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=130,theta2=230,color="black")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)

    #Tidy Axes
    plt.axis('off')

    #Draw passes
    for idx, row in df_passes.iterrows():
        start = row['location']
        end = row['pass_end_location']
        plt.plot([start[0], end[0]], [start[1], end[1]], color="blue")
        plt.plot(start[0], start[1], ".", color="green")
        plt.plot(end[0], end[1], "x", color="green")

    plt.show()
```

```python
first_arg = df_arg_pass[df_arg_pass.index <= 476]    # first half of the match
second_arg = df_arg_pass[df_arg_pass.index > 476]    # second half of the match
show_passes(first_arg)
show_passes(second_arg)
```

```python
first_fra = df_fra_pass[df_fra_pass.index <= 480]    # first half of the match
second_fra = df_fra_pass[df_fra_pass.index > 480]    # second half of the match
show_passes(first_fra)
show_passes(second_fra)
```
